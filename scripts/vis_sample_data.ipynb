```python
import numpy as np
pbs = np.load("../examples/sample_data/PBS.npy")
pbs = pbs[:, 1]
particle = np.load("../examples/sample_data/Particle.npy")
particle = particle[:, 1]

print("pbs.shape", pbs.shape)
print("particle.shape", particle.shape)

# データの可視化
import matplotlib.pyplot as plt

```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# ===== データ切り出し =====
n_points = 1000

pbs_slice = pbs[:n_points]
particle_slice = particle[:n_points]

# ===== カラーマップ =====
cmap = cm.get_cmap("cividis")
color_pbs = cmap(0.2)
color_particle = cmap(0.8)

# ===== プロット =====
plt.figure(figsize=(12, 5))

time_axis = np.arange(n_points)

plt.plot(time_axis, pbs_slice, label="PBS", linewidth=3, color=color_pbs)
plt.plot(time_axis, particle_slice, label="Particle", linewidth=3, color=color_particle)

# ===== ラベル =====
plt.xlabel("Time", fontsize=16)
plt.ylabel("Amplitude", fontsize=16)
plt.title("Time Series (First 1000 Points)", fontsize=18)

# ===== その他 =====
plt.legend(fontsize=14)
plt.grid(True, alpha=0.3, linestyle="--")
plt.tick_params(axis='both', labelsize=14)

plt.tight_layout()
plt.show()
```
